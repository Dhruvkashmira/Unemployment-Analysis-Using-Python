##LABOUR PARTICIPATION RATE AND UNEMPLOYMENT RATE IN EACH MONTH

```python
import plotly.graph_objects as go
df_EE = df.groupby('Month')[['Estimated Unemployment Rate (%)','Estimated Employed','Estimated Labour Participation Rate (%)']].mean()
df_EE = pd.DataFrame(df_EE).reset_index()
month = df_EE['Month']
unemployment_rate = df_EE['Estimated Unemployment Rate (%)']
labour_participation_rate = df_EE['Estimated Labour Participation Rate (%)']

fig = go.Figure()
fig.add_trace(go.Bar(x= month,y=unemployment_rate,name='unemployment_rate'))
fig.add_trace(go.Bar(x= month,y=labour_participation_rate,name='labour_participation_rate'))
fig.update_layout(title='UNEMPLOYMENT RATE AND LABOUR PARTICIPATION RATE IN EACH MONTH',xaxis_title='Month',yaxis_title='Rate')
fig.show()
```

```python
import plotly.graph_objects as go
df_EE_2020 = df_2020.groupby('Month')[['Estimated Unemployment Rate (%)','Estimated Employed','Estimated Labour Participation Rate (%)']].mean()
df_EE_2020= pd.DataFrame(df_EE_2020).reset_index()
month = df_EE_2020['Month']
unemployment_rate = df_EE_2020['Estimated Unemployment Rate (%)']
labour_participation_rate = df_EE_2020['Estimated Labour Participation Rate (%)']

fig = go.Figure()
fig.add_trace(go.Bar(x= month,y=unemployment_rate,name='unemployment_rate'))
fig.add_trace(go.Bar(x= month,y=labour_participation_rate,name='labour_participation_rate'))
fig.update_layout(title='UNEMPLOYMENT RATE AND LABOUR PARTICIPATION RATE IN EACH MONTH',xaxis_title='Month',yaxis_title='Rate')
fig.show()
```
